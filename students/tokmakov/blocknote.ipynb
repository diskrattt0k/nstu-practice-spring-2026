```python
from students.tokmakov.lesson2 import Exercise
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
x = data.data
y = data.target
y = (y == 0).astype(int)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
model = Exercise.create_logistic_model(x.shape[1])
hyperparams = Exercise.get_iris_hyperparameters()
Exercise.fit(
    model,
    x_train,
    y_train,
    lr=hyperparams["lr"],
    n_iter=4,
    batch_size=int(hyperparams["batch_size"]),
)
```

```python
print("Train accuracy:", model.metric(x_train, y_train))
print("Train accuracy test:", model.metric(x_test, y_test))
print("Loss train:", model.loss(x_train, y_train))
print("Loss test:", model.loss(x_test, y_test))
```
